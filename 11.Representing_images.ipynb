```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img_bgr = cv2.imread('data/lena.jpg')
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(121)
plt.imshow(img_bgr)
plt.subplot(122)
plt.imshow(img_rgb);
```

```python
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
corners = cv2.cornerHarris(img_gray, 2, 3, 0.04)
```

```python
plt.figure(figsize=(12, 6))
plt.imshow(corners, cmap='gray')
```

```python
orb = cv2.ORB_create()
kp, des = orb.detectAndCompute(img_bgr, None)
```

```python
img_kp = cv2.drawKeypoints(img_rgb, kp, None, flags=cv2.DrawMatchesFlags_DRAW_RICH_KEYPOINTS)
```
